```python
from matplotlib.backends.backend_agg import FigureCanvasAgg #import plotting on canvas
from matplotlib.figure import Figure
fig=Figure()
#associate fig with background
canvas=FigureCanvasAgg(fig)
ax=fig.add_subplot(111) #pass 3 things
#plot th point(3,2)
ax.plot(3,2,'.')
#save the figure to test.png
canvas.print_png('test.png')
```

```python
import numpy as np
x=np.array([1,2,3,4,5,6,7,8])
y=x
colors=['green']*(len(x)-1)
colors.append('red')
plt.figure()
#plt.figure

plt.scatter(x,y, s=100,c=colors)
```
